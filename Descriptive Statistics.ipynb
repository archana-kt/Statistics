# Normal Distribution

```python
import matplotlib.pyplot as plt
```

```python
# Generate 1000 random samples from a normal distribution with mean 10 and standard deviation 2

x = np.random.normal(10, 2, size=1000)
```

```python
x
```

```python
# Calculate mean and standard deviation
```

```python
mean = np.mean(x)
```

```python
mean
```

```python
np.sum(x) / len(x)
```

```python
np.average(x)
```

```python
std = np.std(x)
```

```python
std
```

```python
x1 = []
```

```python
for i in x:
    x1.append(i - mean)
```

```python
x1
```

```python
square = 0
```

```python
for i in x1:
    square += i * i
```

```python
square
```

```python
var = square / len(x)
```

```python
var
```

```python
np.var(x)
```

```python
std = np.sqrt(var)
```

```python
std
```

```python
np.std(x)
```

```python
# Create an array of values for the x-axis

x_values = np.linspace(mean - 3*std, mean + 3*std, 100)
```

```python
x_values
```

```python
# Calculate the corresponding y-values using the normal distribution formula

y_values = (1/(std * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x_values - mean)/std)**2)
```

```python
y_values
```

```python
# Plotting the bell curve

plt.plot(x_values, y_values)
plt.show()
```
